```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dadosIC = pd.read_csv('ice_cream_sales.csv')
dadosIC
```

```python
celsius = (dadosIC['Temperature']- 32) * 5/9
dadosIC['Temperature'] = celsius
dadosIC
```

```python
plt.title('Ice Cream Truck')

dadosIC = dadosIC.sort_values(by="Temperature")


plt.plot(dadosIC['Temperature'], dadosIC['Sales'], 'r.--', label='US$')


plt.legend(['US$'])
plt.xlabel('Temperatura (Celsius)')
plt.ylabel('Sales (US$)')


plt.savefig('icecream_sales.png')
plt.savefig('icecream_sales.pdf')

plt.show()
```
